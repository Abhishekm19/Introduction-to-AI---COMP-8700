```python
import math

class TicTacToe:
    def __init__(self):
        self.board = [' ' for _ in range(9)]  # Represents a 3x3 board
        self.current_player = 'X'

    def print_board(self):
        for row in [self.board[i*3:(i+1)*3] for i in range(3)]:
            print('| ' + ' | '.join(row) + ' |')

    def available_moves(self):
        return [i for i, val in enumerate(self.board) if val == ' ']

    def make_move(self, move):
        self.board[move] = self.current_player
        self.current_player = 'O' if self.current_player == 'X' else 'X'

    def winner(self):
        # Check rows, columns, and diagonals for a winner
        lines = [(0, 1, 2), (3, 4, 5), (6, 7, 8),
                 (0, 3, 6), (1, 4, 7), (2, 5, 8),
                 (0, 4, 8), (2, 4, 6)]

        for line in lines:
            if self.board[line[0]] == self.board[line[1]] == self.board[line[2]] != ' ':
                return self.board[line[0]]
        if ' ' not in self.board:
            return 'Tie'
        return None

def minimax(game):
    def max_value(state):
        if game.winner() is not None:
            if game.winner() == 'X':
                return -1
            elif game.winner() == 'O':
                return 1
            else:
                return 0

        v = -math.inf
        for move in game.available_moves():
            game.make_move(move)
            v = max(v, min_value(game))
            game.board[move] = ' '
        return v

    def min_value(state):
        if game.winner() is not None:
            if game.winner() == 'X':
                return -1
            elif game.winner() == 'O':
                return 1
            else:
                return 0

        v = math.inf
        for move in game.available_moves():
            game.make_move(move)
            v = min(v, max_value(game))
            game.board[move] = ' '
        return v

    best_move = -1
    best_val = -math.inf
    for move in game.available_moves():
        game.make_move(move)
        val = min_value(game)
        game.board[move] = ' '
        if val > best_val:
            best_val = val
            best_move = move
    return best_move

# Play Tic-Tac-Toe
def play_game():
    game = TicTacToe()
    while not game.winner():
        game.print_board()
        if game.current_player == 'X':
            move = int(input('Enter your move (0-8): '))
        else:
            move = minimax(game)
        game.make_move(move)

    game.print_board()
    winner = game.winner()
    if winner == 'Tie':
        print("It's a tie!")
    else:
        print(f"{winner} wins!")

play_game()
```
